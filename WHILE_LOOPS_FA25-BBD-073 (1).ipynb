```python
secnum = 69
guess = int(input("guess the number : "))
attempt = 1
while guess != secnum:
    if guess > secnum:
        print("Too High")
    else:
        print("Too Low")
    guess = int(input("Try Again"))
    attempt += 1
print("Correct Guess! Attempts : ", attempt)    
```
